```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as mpp
```

```python
d1 = pd.read_csv(r"E:\ELEVATE LABS\TASK - 2\Titanic-Dataset.csv")
```

```python
d1['Age'].hist(bins = 30, edgecolor = 'Black', color = 'Red')
mpp.title("Distribution of Age") 
mpp.xlabel("Age")
mpp.ylabel("Frequency")
```

```python
d1['Fare'].hist(bins = 30, edgecolor = 'Black', color = 'Green')
mpp.title("Distribution of Fare") 
mpp.xlabel("Fare")
mpp.ylabel("Frequency")
```

```python
sb.boxplot(x=d1['Age'])   
mpp.title("Boxplot of Age")
```

```python
sb.boxplot(x=d1['Fare'], color='orange')
mpp.title("Boxplot of Fare")
```

```python
numeric_d1 = d1.select_dtypes(include=['int64', 'float64'])
```

```python
mpp.figure(figsize=(10,6))
sb.heatmap(numeric_d1.corr(), annot=True, cmap='coolwarm')
mpp.title("Correlation Matrix")
```

```python
sb.pairplot(d1[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived')
mpp.suptitle("Pairplot by Survival", y=1.00)
```

```python
sb.countplot(x='Survived', hue='Sex', data=d1)
mpp.title("Survival Count by Gender")
```
